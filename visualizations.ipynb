```python
# Step 1
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from numpy import pi

qreg_q = QuantumRegister(3)  # this is 3 qubits
qc = QuantumCircuit(qreg_q)  # our simple circuit will only have 3 qubits

state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
# Step 2
qreg_q = QuantumRegister(3)  # this is 3 qubits
qc = QuantumCircuit(qreg_q)  # our simple circuit will only have 3 qubits

qc.x(qreg_q[0])
qc.z(qreg_q[1])
qc.y(qreg_q[2])

state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
# Step 3
qreg_q = QuantumRegister(3)  # this is 3 qubits
qc = QuantumCircuit(qreg_q)  # our simple circuit will only have 3 qubit

for register in qreg_q:
    qc.ry(pi/2, register)
    qc.rz(pi/4, register)

state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
# Step 4
qreg_q = QuantumRegister(3)  # this is 3 qubits
qc = QuantumCircuit(qreg_q)  # our simple circuit will only have 3 qubit

for register in qreg_q:
    qc.ry(pi/2, register)
    qc.rz(pi/4, register)

qc.rx(pi/2, qreg_q[0])
qc.ry(pi/2, qreg_q[1])
qc.rz(pi/2, qreg_q[2])

state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
# Step 5 
qreg_q = QuantumRegister(2) 
qc = QuantumCircuit(qreg_q)

qc.x(qreg_q[1])

state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
# Step 6
qreg_q = QuantumRegister(2) 
qc = QuantumCircuit(qreg_q)

qc.x(qreg_q[1])
qc.h(qreg_q[0])
qc.h(qreg_q[1])

state = Statevector(qc)
plot_bloch_multivector(state)
```
